```python
#########################################################################################################
#set up libraries
import requests
from bs4 import BeautifulSoup
import csv
import re
import datetime
from pandas import DataFrame as df
from time import sleep
import time
import sys
reload(sys)
sys.setdefaultencoding('utf8')
##########################################################################################################
#get soup of each hote;
#set up lists
players = []
tlst = []
weight = []
height = []
name = []
age = []
rank = []
wlratio = []
turpro = []
bday = []
court = []
ctype = []
ct = []
tp = []
country = []
playstyle = []
coach = []
#basic setup to get hotel info
bs_url = 'http://www.atpworldtour.com/en/scores/results-archive'
cty_url = bs_url
response = requests.get(cty_url)
html = response.text.encode('utf-8')
# create a beautifulsoup object of the search page of boston hotel
atp_soup = BeautifulSoup(html, "lxml")
#print(cty_url)
#############################################################################################
#getting court type
for i in range (0,201):#201
    details = atp_soup.find_all('div', class_ = 'item-details')[i].find(text=True)
    details = ' '.join(details.split())
    #print(details)
    court.append(details)
    
    value = atp_soup.find_all('div', class_ = 'item-details')[i].find('span', class_ = 'item-value').find(text=True)
    value = ' '.join(value.split())
    ctype.append(value)
        
court = court[1::3]
ctype = ctype[1::3]
#########################################################################################################
for i in range (0,67):
    tennis_boxes = atp_soup.find_all('a',class_='button-border')[i]['href']
    tlst.append(tennis_boxes)
#############################################################################################
#looping through to get each link
for x in range (0,67):#67
    bs_url = 'http://www.atpworldtour.com'+ tlst[x]
    print(bs_url)
    res = requests.get(bs_url)
    html = res.text.encode('utf-8')
    # create a beautifulsoup object of the search page of boston hotel
    tennis_soup = BeautifulSoup(html, "lxml")
#############################################################################################
    #length of loop calc
    lenply = len(tennis_soup.find_all('td', class_='day-table-name'))
    #player links
    #player name
    for w in range (0,lenply):    
        name.append(tennis_soup.find_all('td', class_='day-table-name')[w].text.strip('\t\r\n'))
    ########################
    for w in range (0,lenply):    
        players_boxes = tennis_soup.find_all('td', class_='day-table-name')[w].find('a',class_ = "")['href']
        players.append(players_boxes)
    ############################################
    for z in range (0, lenply):
        bs2_url = 'http://www.atpworldtour.com'+ players[z]
        #print(bs2_url)
        res2 = requests.get(bs2_url)
        html2 = res2.text.encode('utf-8')
        # create a beautifulsoup object of the search page of boston hotel
        player_soup = BeautifulSoup(html2, "lxml")
        ##############################################################
        #attributes
        for y in  range (0,1):
            ##############################################################
            try:
                bday.append(player_soup.find('span', class_='table-birthday').text.strip('\t\r\n'))
            except:
                bday.append("N/A")
            ##############################################################    
            try:    
                weight.append(player_soup.find('span', class_='table-weight-lbs').text.strip('\t\r\n'))
            except:
                weight.append("N/A")
            ##############################################################    
            try:    
                height.append(player_soup.find('span', class_='table-height-cm-wrapper').text.strip('\t\r\n'))
            except:
                height.append("N/A")
            ##############################################################                   
            try:
                rank.append(player_soup.find('div', class_='data-number').text.strip('\t\r\n'))
            except:
                rank.append("N/A")
            ##############################################################                  
            try:
                age.append(player_soup.find('div', class_='player-profile-hero-table').find('div',class_ = 'table-big-value').text.strip('\t\r\n')[0:2])
            except:
                age.append("N/A")
            ##############################################################
            try:
                turpro.append(player_soup.find_all('div', class_='player-profile-hero-table')[y].find_all('div',class_ = 'table-big-value')[1].text.strip('\t\r\n'))
            except:
                turpro.append("N/A")
            ##############################################################
            try:
                playstyle.append(player_soup.find_all('div', class_='table-value')[2].text.strip('\t\r\n'))
            except:
                playstyle.append("N/A")
            ##############################################################
            try:
                country.append(player_soup.find('div', class_='table-value').text.strip('\t\r\n'))
            except:
                country.append("N/A")
            ##############################################################
            try:
                coach.append(player_soup.find_all('div', class_='table-value')[3].text.strip('\t\r\n'))
            except:
                coach.append("N/A")
            ##############################################################            
            try:
                wlratio.append(player_soup.find_all('div', class_='stat-value')[6].text.strip('\t\r\n'))
            except:
                wlratio.append("N/A")              
            ##############################################################    
##############################################################################################
tot = [name,age,bday,weight,height,rank,turpro,country,playstyle,coach,wlratio]
tennistable2 = df(tot).transpose()
tennistable2.columns = ['Name','Age','Birthday','Weight','Height','Rank','Turned Pro','Country','Playstyle','Coach','Wins:Loss']
##############################################################################################
```
