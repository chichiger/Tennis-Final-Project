## Scraping Match Results for each Tournament ##
#### We are scraping the winner, loser, and the score for each round of every tournament####

```python
#####################################################################################
#set up libraries
import requests
from bs4 import BeautifulSoup
import csv
import re
import datetime
from pandas import DataFrame as df
from time import sleep
import time
#####################################################################################
#basic setup to get tennis info
main_url = 'http://www.atpworldtour.com/en/scores/results-archive'
cty_url = main_url
response = requests.get(cty_url)
html = response.text.encode('utf-8')
# create a beautifulsoup object of the result page of the tennis matches
atp_soup = BeautifulSoup(html, "lxml")
#####################################################################################
#get soup of each tournament;
#set up lists. The lists are used to store the information you scrape and then strip the
#spacing issues that came from the scraping
score = []
players = []
winners = []
losers = []
tlst = []

#loop through each tournament
for i in range (0,67):
    tennis_boxes = atp_soup.find_all('a',class_='button-border')[i]['href']
    tlst.append(tennis_boxes)
#####################################################################################
#looping through to get each link
for x in range (0,67):#67
    bs_url = 'http://www.atpworldtour.com'+ tlst[x]
    print(bs_url)
    res = requests.get(bs_url)
    html = res.text.encode('utf-8')
    # create a beautifulsoup object of the search page of boston hotel
    tennis_soup = BeautifulSoup(html, "lxml")
#######################################################################################
    #length of loop calc
    lenscr = len(tennis_soup.find_all('td', class_='day-table-score'))
    lenply = len(tennis_soup.find_all('td', class_='day-table-name'))
    #scores
    for z in range (0,lenscr):
        score.append(tennis_soup.find_all('td', class_='day-table-score')[z].text.strip('\t\r\n'))
    #players
    for w in range (0,lenply):    
        players.append(tennis_soup.find_all('td', class_='day-table-name')[w].text.strip('\t\r\n'))
    #winners
    winners = players[0::2]
    #losers
    losers = players[1::2]
#########################################################################################
tot = [winners,losers,score]
tennistable = df(tot).transpose()
tennistable.columns = ['winner','loser','match score']
#tennistable.to_csv('tennistest1.csv')
```
