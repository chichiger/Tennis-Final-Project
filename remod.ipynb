```python
import requests
from bs4 import BeautifulSoup
import csv
import re
import datetime
from pandas import DataFrame as df
from time import sleep
import time
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tent = pd.read_csv("matchresults.csv")
```

```python
a = tent['match score'].str.split()
```

```python
tent['Court'][1]
```

```python
s1 = []
s2 = []
s3 = []
s4 = []
s5 = []
s6 = []
crt = []
typ = []
wnr = []
lsr = []
for x in range (0,len(tent)):
    if(a[x][0]):
        s1.append(a[x][0][0])#this might skew the results based on the fact that closer matches would ahve a higher score that matches where the winner sweeps the loser
        crt.append(tent['Court'][x])
        typ.append(tent['Type'][x])
        wnr.append(tent['winner'][x])
        lsr.append(tent['loser'][x])
    else:
        s1.append('0')
    
    if(a[x][0]):
        s2.append(a[x][0][1])#this might skew the results based on the fact that closer matches would ahve a higher score that matches where the winner sweeps the loser
    else:
        s2.append('0')
        
    if(len(a[x])>=2):
        s3.append(a[x][1][0])#this might skew the results based on the fact that closer matches would ahve a higher score that matches where the winner sweeps the loser
    else:
        s3.append('0')
    
    if(len(a[x])>=2):
        s4.append(a[x][1][1])#this might skew the results based on the fact that closer matches would ahve a higher score that matches where the winner sweeps the loser
    else:
        s4.append('0')
        
    if(len(a[x])>=3):
        s5.append(a[x][2][0])#this might skew the results based on the fact that closer matches would ahve a higher score that matches where the winner sweeps the loser
    else:
        s5.append('0')
    
    if(len(a[x])>=3):
        s6.append(a[x][2][1])#this might skew the results based on the fact that closer matches would ahve a higher score that matches where the winner sweeps the loser
    else:
        s6.append('0')
        

fnl = [wnr,lsr,crt,typ,s1,s2,s3,s4,s5,s6]
```

```python
fnl
tt = df(fnl).transpose()
tt.columns = ['winner','loser','court','Type','set1.1','set1.2','set2.1','set2.2','set3.1','set3.2']
```

```python
t2 = pd.read_csv("playerdata.csv")
```

```python
t2
```

```python
a2 = t2['Birthday']
h2 = t2['Height']
w2 = t2['Weight']
n2 = t2['Name']
g2 = t2['Age']
r2 = t2['Rank']
tp2 = t2['Turned Pro']
ct2 = t2['Country']
ps2 = t2['Playstyle'].str.split()
ch2 = t2['Coach'].str.split('and')
wl2 = t2['Wins:Loss'].str.split('-')
```

```python
bdy = []
ht = []
wt = []
nm = []
ag = []
rnk = []
tdp = []
cty = []
hnd = []
bkd = []
wns = []
lss = []
nchs = []

for d in range(0,len(a2)):
            bdy.append(a2[d][1:-1])
            if(type(h2[d]) == str ):
                ht.append(h2[d][1:-3])
            else:
                ht.append(int('0'))
            wt.append(w2[d])
            nm.append(n2[d])
            ag.append(g2[d])
            rnk.append(r2[d])
            tdp.append(tp2[d])
            cty.append(ct2[d])
            #print(d)
            if(type(h2[d]) == str ):
                hnd.append(ps2[d][0][0:-1])
                if(len(ps2[d]) == 3):
                    bkd.append(ps2[d][1]+' '+ps2[d][2])
            else:
                hnd.append('none')
                bkd.append('none')

            wns.append(wl2[d][0])
            lss.append(wl2[d][1])

            if(type(ch2[d]) == list):
                nchs.append(len(ch2[d]))
            else:
                nchs.append(int('0'))

```

```python
rmt = [nm,ag,bdy,ht,wt,rnk,cty,tdp,wns,lss,hnd,bkd,nchs]
```

```python
rf = df(rmt).transpose()
rf.columns = ['Name','Age','Birthday','Height','Weight','Rank','Country Represent','Year Turned Pro','Wins','Losses','Handed','Backhand Style','Number of Coaches']
```

```python
rf.drop_duplicates(subset=['Rank'])
```

```python
rf.to_csv('Final_Player_Data.csv')
```
